```python
for sample in samples:
    for nrand in nrands:
        tpath = rpath.format(sample, nrand)
        for chrom in range(1, 23) + ['X']:
            chrom = str(chrom)
            fh = open(glob(tpath + '{}_EigVect1_*.tsv'.format(chrom))[0],'r')
            next(fh)
            count = 0
            for line in fh:
                count += 1
            if genome[chrom]!=count:
                print('Sample:',sample,'Repl:',nrand,'Chrom:',chrom,'Expected:',genome[chrom],'Observed:',count)
```
